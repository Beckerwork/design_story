    <h5>What is Normalization?</h5>
    Normalization is the process of rescaling values into a fixed range — usually <b>[0, 1]</b>.
    <br><br>
    This is useful because machine learning models often work better when all input features
    are on a similar scale. Without normalization, a feature with large values can dominate
    the model simply because of its magnitude, not its importance.
    <br><br>
    The formula is: &nbsp; <code>(x - min) / (max - min)</code>
  </div>
