## Good to Know
________

#### 1) Rechenzeichen

Da wir uns mit mathematischer Modellierung beschäftigen, müssen wir oft auch mathematische Formeln in das Codefeld eingeben. </br>
In der "Code-Sprache" sehen die <b>Rechenzeichen</b> manchmal ein wenig anders aus:

* <b>Addition</b> " + " : &nbsp; wird im Code ebenfalls mit `+` geschrieben.
* <b>Subtraktion</b> " - " : &nbsp; wird im Code ebenfalls mit `-` geschrieben.
* <b>Multiplikation</b> " $\cdot$ " : &nbsp; wird im Code mit `*` geschrieben.
* <b>Division</b> " : " : &nbsp; wird im Code mit `/` geschrieben.
* <b>Potenzen</b> " $x^{2}$ ": &nbsp; wird im Code mit `^` geschrieben, also z.B. `x^2`.

######    

#### 2) Shortcuts

Manchmal kommt es vor, dass du dir viel <b>Tipparbeit sparen</b> kannst, wenn du einen Teil deines bereits geschriebenen Codes kopierst und woanders wieder einfügst. Dafür gibt es sogenannte <b>Tastatur-Shortcuts</b>:

* Markiere den gewünschten Codeabschnitt im Codefeld mit deinem Mauszeiger.
* Drücke dann auf der Tastatur `Strg+C` (Eselsbrücke: "C" steht für "copy").
* Klicke an die Stelle im Codefeld, wo du den kopierten Abschnitt einfügen möchtest.
* Drücke dann auf der Tastatur `Strg+V`. 

Um Codefelder auszuführen kannst du die Tastenkombination `Strg+Enter` verwenden.

Um deine letzte Eingabe rückgängig zu machen kannst du `Strg+Z` verwenden.

<i class="fa fa-exclamation-triangle" style="font-size:20px"></i>&nbsp; <b>Hinweis für Mac-User:</b> &nbsp; Statt der Taste `Strg` wird hier `Command`verwendet!

######  

#### 3) Fehlermeldungen

Es kann passieren, dass du beim Ausführen eines Codefelds keine schöne Ausgabe, sondern eine <font color="firebrick">rot hinterlegte</font> <b>Fehlermeldung</b> erhältst.</br>Die häufigsten davon sind:

* <b>UndefVarError</b>: &nbsp; Hier hast du vermutlich weiter oben vergessen, ein Codefeld auszuführen. Überprüfe daher, ob du auf dem bisherigen Arbeitsblatt alle Codefelder der Reihe nach ausgeführt hast. Die Fehlermeldung sieht dann in etwa so aus:

<div style="text-align:left">
<img src="../figs/Fehlermeldung1.png" width="700", class="center">
</div> 

* <b>MethodError</b>: &nbsp; Hier handelt es sich vermutlich um einen Tippfehler in deinem Code. Überprüfe deine Eingabe (z.B. Komma statt Punkt, falsche/fehlende Klammer, Schreibfehler, ...). Die Fehlermeldung sieht dann in etwa so aus:

<div style="text-align:left">
<img src="../figs/Fehlermeldung2.png" width="700", class="center">
</div> </br>

<i class="fa fa-exclamation-triangle" style="font-size:20px"></i>&nbsp; <b>Hinweis:</b> &nbsp; Nachdem du deinen Fehler korrigiert hast, musst du das Codefeld mit der Fehlermeldung <b>erneut ausführen</b>, um die richtige Ausgabe zu erhalten.
